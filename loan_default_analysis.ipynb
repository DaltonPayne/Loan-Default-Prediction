```python
#Load the dataset and print the first 5 rows
import pandas as pd
import numpy as np

df = pd.read_csv('Loan_default.csv')
df.head()
```

```python
#Check the shape of the dataset
print(f"The dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")

#Summary statistics
df.describe(include='all')
```

```python
#Check for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Custom color palette for the bars
palette = sns.color_palette("Set2", len(df['Default'].unique()))  # Creates a color palette for each unique value in 'Default'

# Plot the distribution of the 'Default' variable
sns.countplot(x='Default', data=df, palette=palette)
plt.title('Distribution of Loan Default')
plt.show()

# Distribution of 'Default' in percentages
df['Default'].value_counts(normalize=True) * 100

```

```python
#Check data types
df.dtypes
```

```python
#Convert data types as necessary
numeric_columns = ['Age', 'Income', 'LoanAmount', 'CreditScore', 'MonthsEmployed',
                   'NumCreditLines', 'InterestRate', 'LoanTerm', 'DTIRatio', 'Default']

for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
#List of categorical columns
categorical_columns = ['Education', 'EmploymentType', 'MaritalStatus', 'HasMortgage',
                       'HasDependents', 'LoanPurpose', 'HasCoSigner']

#Convert to category data type
for col in categorical_columns:
    df[col] = df[col].astype('category')
```

```python
#Plotting categorical variables
for col in categorical_columns:
    sns.countplot(x=col, data=df, hue='Default', palette=palette)  # Adding palette here for hue-based color
    plt.title(f'Distribution of {col} by Default')
    plt.xticks(rotation=45)
    plt.show()
```

```python
#Default rate by categories
for col in categorical_columns:
    default_rate = df.groupby(col)['Default'].mean() * 100
    default_rate.plot(kind='bar', color=sns.color_palette("Set2", len(default_rate)))
    plt.title(f'Default Rate by {col}')
    plt.ylabel('Default Rate (%)')
    plt.xticks(rotation=45)
    plt.show()
```

```python
from matplotlib.colors import LinearSegmentedColormap

# Define a custom orange-green colormap for better representation of negative correlations
orange_green_cmap = LinearSegmentedColormap.from_list("OrangeGreen", ['#ff4500', '#FFFFFF', '#32CD32'])  # Darker orange to white to green

# Plot the correlation matrix with the improved custom colormap
plt.figure(figsize=(12, 10))
corr = df[numeric_columns].corr()  # Use the correct list of numerical columns
sns.heatmap(corr, annot=True, cmap=orange_green_cmap, center=0, vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```
